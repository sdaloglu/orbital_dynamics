```python
# This script uses the imageio module to create 
# an animated gif from a series of png images
import imageio
import os

# note that in order for the 'sorted()' to properly sort files with numbered 
# filenames like 'snapshot_nn.png', you will need to pad the numbers with zero
# so they all have the same number of digits (ie, 01, 02, ... 10, 11, ...):
n=3
print(f'{n:02d}')
```

```python
# this assumes that the png image files are all in a directory called 
# 'test_snaps' and have filenames that allow them to be sorted 
# (ie they are named something like 'snapshot_nn.png')
png_dir = './test_snaps'
movie_filename = 'test_movie_1.gif'

images = []
for file_name in sorted(os.listdir(png_dir)):
    if file_name.endswith('.png'):
        file_path = os.path.join(png_dir, file_name)
        print(file_name, file_path)
        images.append(imageio.imread(file_path))
imageio.mimsave(f'{png_dir}/{movie_filename}', images, fps=10)
```
